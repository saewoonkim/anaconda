```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
import sys
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

start_time = time.perf_counter()
chrome_options = Options()
chrome_options.add_argument('--start-maximized')
## 옵션을 사용하기 위해서 아규먼트를만듬 웹사이트최대화를시키면 크롤링이더잘됨 
CoffeeBean_URL = 'https://www.coffeebeankorea.com/store/store.asp'
wd = webdriver.Chrome(options=chrome_options)
##창 최대환 시킨 옵션을 적용시켜주겠다는 명령어
wd.get(CoffeeBean_URL)
#for 문 밖에 나와있다는것은 사이트를 한번 방문만 하게하는명령어
wd.implicitly_wait(5)
##실제로는 60초 정도 주는게 좋다 기다리는시간을 정해준다
##최대 5초까지 기다리라는 명령어

def CoffeeBean_store(result):
    for i in range(1,225): 
        try:
            wd.find_element(By.CSS_SELECTOR, f"#storeListUL > li:nth-child({i}) > div.num").click() 
            ##셀러늄 함수 s 가 붙어있는거는 요소가 많을때 요소가많이없으면 t 만 붙이면됨
            store_nam = wd.find_element(By.CSS_SELECTOR, 'div.title').text
            store_add = wd.find_element(By.CSS_SELECTOR, 'div.ellipsis').text
            store_tel = wd.find_element(By.CSS_SELECTOR, 'div.jibun.ellipsis').text
    
            print(store_nam)
            print(store_add)
            print(store_tel)
            result.append([store_nam]+[store_add]+[store_tel])
        except:
            sys.stdout.writelines("error")
            continue
    return 

result = []
print('coffeebean crawling')
CoffeeBean_store(result)
CB_tbl = pd.DataFrame(result, columns = ('store', 'address', 'phone'))
##CB_tbl.to_csv('coffeebean7.csv', encoding='euc-kr', mode = 'w', index=True)
##CB_tbl.to_csv('coffeebean5.csv', encoding='cp949', mode = 'w', index=True)
CB_tbl.to_csv('coffeebean6.csv', encoding='utf-8-sig', mode = 'w', index=True)



end_time = time.perf_counter()
print("경과시간 : ", end_time - start_time)  
## 코드 실행시간 측정   ### 22초 ~ 73초 소요 : (매장 224개 매장정보 모두 가지고 옴)

wd.quit()
sys.exit("크롤링 종료!!")
```
