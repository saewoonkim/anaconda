```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df_class = df.groupby('class').agg(mean_age=('age', 'mean'))
```

```python
df_class
```

```python
df_hclass = df.query("alive == 'yes'")
```

```python
df_hclass
```

```python
df_hclass.groupby('class').agg(alive=('class','count')).plot.bar(rot=0)
```

```python
df['Family'] = df['sibsp'] + df['parch'] + 1
```

```python
df
```

```python
sns.countplot(x = 'Family', hue = 'survived', data = df)
```

```python
df_survived = df.groupby('sex').agg(mean_survived = ('survived','mean'))
```

```python
df_survived.plot.bar(rot=0)
```
